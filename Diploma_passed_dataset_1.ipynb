```python
from rdkit import Chem, DataStructs
from rdkit.Chem import Descriptors, Draw, AllChem
from rdkit.Chem.SaltRemover import SaltRemover
```

```python
# importing sdf
molecules = Chem.SDMolSupplier('./sdf/diploma_dataset_1_passed_mols.sdf')
```

```python
a_tub_side_c_passed_mols = [mol for mol in molecules if float(mol.GetProp('a_tub_side_c_passed_mols')) >= 0.85]
b_tub_side_d_passed_mols = [mol for mol in molecules if float(mol.GetProp('b_tub_side_d_passed_mols')) >= 0.85]
b_tub_side_g_passed_mols = [mol for mol in molecules if float(mol.GetProp('b_tub_side_g_passed_mols')) >= 0.85]
```

```python
remover = SaltRemover(defnData="[Cl,Br]")
a_tub_side_c_passed_mols[0] = remover.StripMol(a_tub_side_c_passed_mols[0])
b_tub_side_d_passed_mols[0] = remover.StripMol(b_tub_side_d_passed_mols[0])
```

```python
a_tub_side_c_legends = [str(i) for i in range(1, 4)]
b_tub_side_d_legends = [str(i) for i in range(4, 14)]
b_tub_side_g_legends = [str(i) for i in range(14, 16)]
```

```python
Draw.MolsToGridImage(a_tub_side_c_passed_mols, legends=(a_tub_side_c_legends), subImgSize=(250,250))
```

```python
Draw.MolsToGridImage(b_tub_side_d_passed_mols, legends=(b_tub_side_d_legends), subImgSize=(250,250))
```

```python
Draw.MolsToGridImage(b_tub_side_g_passed_mols, legends=(b_tub_side_g_legends), subImgSize=(250,250))
```
